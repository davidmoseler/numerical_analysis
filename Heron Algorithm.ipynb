```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Iteractively approximate square root of 2
# with the Heron formula: x_n = 1/2*(x_(n-1) + 2/x_(n-1))

def estimate(x,n):
    for i in range(n):
        x = (1/2)*(x + 2/x)
    return x

def error(x,n):
    return np.linalg.norm(estimate(x,n) - np.sqrt(2))

error = np.vectorize(error)
```

```python
plt.xlabel("Number of iteractions")
plt.ylabel("Log error")

plt.plot(np.log(error(0.5, np.arange(1,10))))
```

```python
#Convergence is reached in the sixth iteration
# The error is exactly the minimum discernible value for the floating point system.

print(error(0.5, 6))
print(np.finfo(np.float).eps)
```
