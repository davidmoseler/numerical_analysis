```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Tries to approximate arctan'(a) (the derivative of arctan at point a)
# by (arctan(a+h) - arctan(a))/h.
# Since lim h->0 (arctan(a+h) - arctan(a))/h = arctan'(a), this should
# theoretically converge. But floating point has accuracy limitations which
# will make the aproximation worse with very small values of h.

def approx(a,h):
    return (np.arctan(a+h) - np.arctan(a))/h

#Calculates the value comparing the approximation with the true value,
# given by calculus formula arctan'(a) = 1/(1+aÂ²)

def error(a,h):
    return np.linalg.norm(approx(a,h) - 1/(1+a**2))

error = np.vectorize(error)
```

```python
a = 1
h = 10.0**(-np.arange(1,25))

plt.figure(figsize=(10,5))
plt.xlabel("Negative power of 10 for h")
plt.ylabel("Log error")
plt.plot(np.log(error(a, h)))
```
